```python
def classify(i):
    if i == 1:
        return "能源(Energy Sources)"
    elif i == 2:
        return "建筑材料(Building Materials)"
    elif i == 3:
        return "建材运输(Material Transportation)"
    elif i == 4:
        return "碳汇(Carbon Sink)"
    else:
        return "未找到该种类材料名称"
    
def display(i):
    if i == 1:
        return 10
    elif i == 2:
        return 6
    elif i == 3 or 4:
        return 4
    
```
