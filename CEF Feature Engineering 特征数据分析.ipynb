```python
## 读取表格数据
## Read all sheets from the Excel file
xlsx = pd.read_excel('Carbon Emission Factor Database.xlsx', sheet_name=None)

## Access each DataFrame using the sheet name
## df1 = xlsx['能源(Energy Sources)']
df = xlsx['建筑材料(Building Materials)']
## df3 = xlsx['建材运输(Material Transportation)']
## df4 = xlsx['碳汇(Carbon Sink)']
```

```python
## 首先分析建筑材料的碳排放决定因素
df.head
```

```python
## 1.价格CNY/t

df1 = df[df['碳排放因子'] <= 1000]
df1 = df1[df1['价格CNY/t'] <= 20000]

price = df1.iloc[:, 1]
emission_factor = df1.iloc[:, 5]

plt.scatter(price, emission_factor)
plt.xlabel("Material's Price (CNY/ton)")
plt.ylabel("Carbon Emission Factor (kg CO\u2082e/m\u00B3)")
plt.title("1.Material's Price vs Carbon Emission Factor")
plt.show()
```

```python
## 2.强度MPa

df2 = df[df['碳排放因子'] <= 4000]

intensity = df2.iloc[:, 2]
emission_factor = df2.iloc[:, 5]

plt.scatter(intensity, emission_factor)
plt.xlabel("Material's Intensity (MPa)")
plt.ylabel("Carbon Emission Factor (kg CO\u2082e/m\u00B3)")
plt.title("2.Material's Intensity vs Carbon Emission Factor")
plt.show()
```

```python
## 3.热传导W/(m·K)

df3 = df[df['碳排放因子'] <= 4000]
df3 = df3[df3['热传导W/(m·K)'] <= 3]

heatcap = df3.iloc[:, 3]
emission_factor = df3.iloc[:, 5]

plt.scatter(heatcap, emission_factor)
plt.xlabel("Material's Heat Capacity (W/(m·K))")
plt.ylabel("Carbon Emission Factor (kg CO\u2082e/m\u00B3)")
plt.title("3.Material's Heat Capacity vs Carbon Emission Factor")
plt.show()
```

```python
## 4.密度g/cm³

df4 = df[df['碳排放因子'] <= 4000]
df4 = df4[df4['密度g/cm³'] <= 4]

density = df4.iloc[:, 4]
emission_factor = df4.iloc[:, 5]

plt.scatter(density, emission_factor)
plt.xlabel("Material's Density (g/cm³)")
plt.ylabel("Carbon Emission Factor (kg CO\u2082e/m\u00B3)")
plt.title("Material's Density vs Carbon Emission Factor")
plt.show()
```
